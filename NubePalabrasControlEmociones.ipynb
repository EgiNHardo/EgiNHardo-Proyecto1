```python
# ▶️ Instalar WordCloud (si no está instalada)
!pip install wordcloud

# ▶️ Importar librerías
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from IPython.display import Image, display, FileLink
import random

# ▶️ Lista de palabras
palabras = [
    "Sentimientos", "Afectividad", "Empatía", "Alegría", "Tristeza", "Ira", "Miedo", "Amor",
    "Euforia", "Frustración", "Llanto", "Risa", "Lenguaje corporal", "Reacciones emocionales",
    "Expresión emocional", "Juicio", "Razonamiento", "Evaluación", "Análisis", "Alternativas",
    "Elección", "Criterio", "Deliberación", "Resolución", "Planificación", "Intuición", "Duda",
    "Preferencia", "Estrategia", "Priorización", "Conclusión", "Autonomía", "Responsabilidad",
    "Valores", "Perspectiva", "Vulnerabilidad", "Catarsis", "Regulación emocional", "Autocontrol",
    "Disciplina", "Moderación", "Templanza", "Paciencia", "Refrenar", "Tolerancia a la frustración",
    "Postergación de la gratificación", "Inhibición", "Regulación conductual", "Impulsividad",
    "Conciencia emocional", "Autoconciencia", "Dominio propio", "Estabilidad emocional",
    "Responsabilidad", "Prudencia", "Reflexión", "Autorregulación", "Comunicación no verbal", "Desahogo"
]

# ▶️ Generar frecuencias aleatorias para estética variada
frecuencias = {palabra: random.randint(10, 100) for palabra in palabras}

# ▶️ Crear la nube de palabras
nube = WordCloud(
    width=1000,
    height=700,
    background_color='white',
    colormap='rainbow',
    prefer_horizontal=0.7,
    max_words=150,
    relative_scaling=0.5,
    collocations=False,
    contour_color='black',
    contour_width=1
).generate_from_frequencies(frecuencias)

# ▶️ Mostrar en pantalla
plt.figure(figsize=(16, 10))
plt.imshow(nube, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
plt.show()

# ▶️ Guardar imagen
output_file = "nube_emociones.png"
nube.to_file(output_file)

# ▶️ Enlace de descarga
print("✅ Imagen guardada. Haz clic para descargar:")
display(FileLink(output_file))
```
